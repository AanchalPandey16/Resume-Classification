as we have 26 doc files that python wont support coz its the older version so we need to convert it to .docx

```python
import os
import win32com.client

base_folder = r"C:\Users\AANCHAL\Desktop\ExcelR\Excelr Project 2\data"

word = win32com.client.Dispatch("Word.Application")
word.Visible = False # make word run in bg

converted_count = 0 # keep count of how many files were succesfully converted

for root, dirs, files in os.walk(base_folder): # recursively scans nested folders
    for file in files: # iterate over each file
        if file.lower().endswith(".doc") and not file.lower().endswith(".docx"):
            
            doc_path = os.path.join(root, file)
            docx_path = doc_path + "x"   # converts file.doc → file.docx

            try:
                doc = word.Documents.Open(doc_path) # open file in word
                doc.SaveAs(docx_path, FileFormat=16)  # 16 = .docx format
                doc.Close()
                converted_count += 1
                print("Converted:", file)

            except Exception as e:
                print("Error converting:", file)

word.Quit()

print(f"\nTotal converted: {converted_count}")
```

```python
import os

base_folder = r"C:\Users\AANCHAL\Desktop\ExcelR\Excelr Project 2\data"

deleted = 0

for root, dirs, files in os.walk(base_folder):
    for file in files:
        if file.lower().endswith(".doc") and not file.lower().endswith(".docx"):
            os.remove(os.path.join(root, file))
            deleted += 1

print("Deleted .doc files:", deleted)
```
